```python
import numpy as np
class LinearRegressionLMS:
    def __init__(self,lr = 0.01, epochs = 1000):
        self.lr = lr
        self.epochs = epochs
        self.theta = None

    def add_bias(self,X):
        ones = np.ones((X.shape[0],1))
        return np.hstach((ones,X))

    def fit(self,X,y):
       X = self.add_bias(X)
       n_samples, n_features = X.shape
       self.theta = np.zeros(n_features)

       for epoch in range(self.epochs):
        for i in range(n_samples):
            xi = X[i]
            yi = y[i]

            y_pred = np.dot(self.theta, xi)
            error = yi - y_pred
            #****************** imp: LMS Update here **********************
            self.theta = self.theta + self.lr * error *xi
            #****************** imp: LMS Update here **********************
    
    def pred(self,X):
        X = self.add_bias(X)
        return np.dot(X,self.theta)

    


       
```

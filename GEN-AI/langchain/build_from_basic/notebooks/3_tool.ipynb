## TOOLS

if the <font color = '#ADD8E6'> <b>LLM is the solider</b></font> then the <font color = '#ADD8E6'><b>tools are the weapons</b></font> like guns, granites, etc...

* tools are callable functions.
* with well-defined inputs and outputs that get passed to a chat model.
* The model decides 
    - when to invoke a tool based on the conversation context
    - what input arguments to provide.

Tools are the python funtions.

Tools basic structure:
1. tool name
2. tool description



### Creation of tools

Tool is created by using the <font color = '#ADD8E6'>@tool</font> decorator which is imported from <font color = '#ADD8E6'>from langchain.tools import tool</font> 

```python
from langchain.tools import tool

@tool
def add(int a, int b) -> int:
    """
    add the integers
    args:
        a: integer
        b: integer
    return:
        c: addition of a and b
    """
    return a+b

# in the function definition we mentioning the types of varibles.
# (int a, int b) are called type hinting.
```

### Custom Tool Name and Tool Description

```python
@tool("calculator", description="Performs arithmetic calculations. Use this for any math problems.")
def calc(expression: str) -> str:
    """Evaluate mathematical expressions."""
    return str(eval(expression))
```

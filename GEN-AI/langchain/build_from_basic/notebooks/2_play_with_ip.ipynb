#### LANGCHAINS STANDARD MESSAGES

👉 In LangChain, messages are the basic unit of communication with an LLM.

They store:

* role: Who is speaking 
* content: What they say
* metadata: Extra info 

___________________________________________________________________________________________
<br> Types of messages:
1. Human message: input from the user
2. System message: instructing the model.
3. AI Assistant: response from the model
4. Tool Messages: Used when AI calls tools

___________________________________________________________________________________________
<br>Simple conversation:

- System: You are a helpful teacher
- User: What is AI?
- AI: AI means machines that can think like humans

```python
from langchain.messages import HumanMessage, AIMessage, SystemMessage

sys_message = SystemMessage("You are a helpfull teacher.")
human_message = HumanMessage("what is ai?")
ai_msg = AIMessage("AI means machine that can think like humans.")
human_message = HumanMessage("how machine can think like humans")
```

```python
messages = [sys_message,human_message,ai_msg,human_message]
```

```python
model.invoke(messages)
```

#### DICTIONARY FORMAT

```python
messages = [
    {"role": "system", "content": "You are a poetry expert"},
    {"role": "user", "content": "Write a haiku about spring"},
    {"role": "assistant", "content": "Cherry blossoms bloom..."}
]
response = model.invoke(messages)
```
